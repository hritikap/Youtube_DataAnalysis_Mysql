### What are the top 15 most liked and commented videos?

```python
%%sql
select Title,Video_id,keyword,likes from videos_stats order by likes desc limit 10;
```

```python
%%sql
select Title,Video_id,keyword,comments from videos_stats order by comments desc limit 10;
```

### How many total views and like does each category have ?

```python
%%sql
select keyword as category, sum(views) as total_views,sum(likes)as total_likes from videos_stats group by keyword;
```

## Lets join the two tables into one and perform more analysis using the newly joined table

```python
%%sql
create table joined as
select videos_stats.Title,videos_stats.Video_Id,videos_stats.Published_at,videos_stats.Keyword,videos_stats.Likes,videos_stats.Comments as total_comments,videos_stats.Views,Comment_data.Comment,Comment_data.Likes as commentslike,Comment_data.Sentiment
from videos_stats
left join Comment_data
On videos_stats.Video_Id=Comment_data.Video_id;
```

```python
%sql select * from joined limit 20
```

### What are the top 10 most-liked comments?

```python
%sql select Video_Id,Comment,commentslike from joined order by commentslike  desc limit 10;
```

### What is the average sentiment score in each keyword category?

```python
%%sql
select keyword,avg(Sentiment) as averagesentimentscore from joined group by keyword;
```

### How many times do company names (i.e., apple or google) appear in each keyword category?

```python
%%sql
select keyword,count(keyword) as No_of_company_apperances  from joined  where keyword='apple'or keyword='google' group by keyword;
```

### How many total videoes does each category have?

```python
%%sql
select keyword,count(Video_Id) as total_videos_per_eachcategory from joined group by keyword;
```
