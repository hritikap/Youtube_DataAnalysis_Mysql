### For this first we read the csv files using the python library "pandas" ansd store it into a seperate dataframe as shown below

```python
import pandas as pd
```

```python
df=pd.read_csv('D:\\Data analyst\\youtubedata\\datasets\\videos-stats.csv')
df
```

### Then we clean the datasets inorder to remove any missing,null or duplicate values from the dataframe

```python
df=df.fillna(0)
df
```

```python
df=df.drop_duplicates(subset='Video_ID')
df
```

```python
df2=pd.read_csv('D:\\Data analyst\\youtubedata\\datasets\\comments.csv')
df2
```

```python
df2=df2.fillna(0)
df2
```

```python
df2=df2.drop_duplicates(subset=['Video_id','Comment'])
df2
```

### Then we create a new sql table for the first dataframe (df1) and insert all the dataframe values in the newly created sql table as shown below:

```python
cursor.execute("""CREATE TABLE videos_stats(Title VARCHAR(200),
               Video_Id VARCHAR(50),
               Published_at DATE,
               Keyword VARCHAR(50),
               likes FLOAT,
               Comments FLOAT,
               Views FLOAT
              )""")
conn.commit()
```

```python
for row in df.itertuples():
    cursor.execute("""insert into videos_stats(Title,Video_Id,Published_at,Keyword,likes,Comments,Views)VALUES(%s,%s,%s,%s,%s,%s,%s)""",
                  (row.Title,row.Video_ID,row.Published_At,row.Keyword,row.Likes,row.Comments,row.Views)
                  )
    
    conn.commit()
```

### Similarly we create another table for the second dataframe(df2) and insert all its values in the new table as shown below:

```python
cursor.execute("""create table Comment_data(
Video_id Varchar(50),
Comment longtext,
Likes int,
Sentiment int
)""")
```

```python
for row in df2.itertuples():
    cursor.execute("""insert into Comment_data(Video_id,Comment,Likes,Sentiment)Values(%s,%s,%s,%s)""",(row.Video_id,row.Comment,row.Likes,row.Sentiment))
    
conn.commit()
```
